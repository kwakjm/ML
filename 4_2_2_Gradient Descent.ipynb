```python
import numpy as np

np.random.seed(42)
x = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

# 학습률
learning_rate = 0.001
# 최대 반복 횟수
max_iterations = 1000

# 초기 기울기와 절편 설정
slope = 0
intercept = 0

# 경사 하강법 수행
for i in range(max_iterations):
    # 예측값 계산
    predicted = slope * x + intercept
    # 오차 계산
    error = predicted - y
    # 기울기와 절편의 변화율 계산
    slope_gradient = np.mean(error * x)
    intercept_gradient = np.mean(error)
    # 기울기와 절편 업데이트
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient
    
    # 기울기와 절편의 변화 출력
    print(f"Iteration {i+1}: 기울기={slope}, 절편={intercept}")

# 최종 기울기와 절편 출력
print("최종 기울기:", slope)
print("최종 절편:", intercept)
```

```python
import numpy as np

np.random.seed(42)
x = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

# 학습률
learning_rate = 0.001
# 최대 반복 횟수
max_iterations = 1000

# 초기 기울기와 절편 설정
slope = 0
intercept = 0

# 경사 하강법 수행
for i in range(max_iterations):
    # 예측값 계산
    predicted = slope * x + intercept
    # 오차 계산
    error = predicted - y
    # 에러율 계산
    error_rate = np.mean(np.abs(error))
    # 기울기와 절편의 변화율 계산
    slope_gradient = np.mean(error * x)
    intercept_gradient = np.mean(error)
    # 기울기와 절편 업데이트
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient
    
    # 기울기와 절편의 변화 및 에러율 출력
    print(f"Iteration {i+1}: 기울기={slope}, 절편={intercept}, 에러율={error_rate}")

# 최종 기울기와 절편 출력
print("최종 기울기:", slope)
print("최종 절편:", intercept)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

class GradientDescentLinearRegression:
    def __init__(self, learning_rate=0.001, epoch=10):
        self.learning_rate = learning_rate
        self.epoch = epoch
        self.a = 0  # 초기 기울기
        self.b = 0  # 초기 절편

    def fit(self, x, y):
        n = x.shape[0]

        for i in range(self.epoch):
            b_gradient = -2 * np.sum(y - (self.a * x + self.b)) / n
            a_gradient = -2 * np.sum(x * (y - (self.a * x + self.b))) / n

            self.b -= self.learning_rate * b_gradient
            self.a -= self.learning_rate * a_gradient

            err = np.sum((y - (self.a * x + self.b)) ** 2) / n  # 손실 계산 위치 수정
            print('loss =', err, 'a =', self.a, 'b =', self.b)

            if i % (self.epoch // 10) == 0:
                plt.scatter(x, y)
                plt.plot(x, self.a * x + self.b)
                plt.show()

    def predict(self, x):
        return self.a * x + self.b
    
    def partial_fit(self, x, y):
        n = x.shape[0]

        for i in range(self.epoch):
            b_gradient = -2 * np.sum(y - (self.a * x + self.b)) / n
            a_gradient = -2 * np.sum(x * (y - (self.a * x + self.b))) / n

            self.b -= self.learning_rate * b_gradient
            self.a -= self.learning_rate * a_gradient
            err = np.sum((y - (self.a * x + self.b)) ** 2) / n  # 손실 계산 위치 수정
            print('loss =', err, 'a =', self.a, 'b =', self.b)

model = GradientDescentLinearRegression(learning_rate=0.001, epoch=100)
model.fit(x, y)

print('\n','추가적인 학습','\n')



model.partial_fit(x, y)

plt.scatter(x, y)
plt.plot(x, model.predict(x))
plt.show()
```
